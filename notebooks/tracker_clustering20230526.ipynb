```python
groups = data.groupby(data.columns[0])
tracker01_id = list(groups.groups.keys())[0]
tracker01 = groups.get_group(tracker01_id)
tracker01.info()
```

```python

# find first timestamp: either smallest receptiontimestamp or measurementtimestamp 
# find last timestamp
# reindex with timeseries from first timestamp to last timestamp -> only day as resolution 
# can there be multiple receptionstimestamps for the same day? 
# order in the existing rows by receiptiontimestamp (need to match the day of the created index), 
#    if none is found -> leave row blanl
```

```python
start_date = min(tracker01["measurementtimestamp"].min(),tracker01["receptiontimestamp"].min()).date()
end_date = max(tracker01["measurementtimestamp"].max(),tracker01["receptiontimestamp"].max()).date()
time_series = pd.date_range(start=start_date, end=end_date, freq='D')
check_days = tracker01["measurementtimestamp"]
df = pd.DataFrame({"date_counter":time_series})
# df[tracker01.columns] = np.nan # pd.DataFrame(index=range(columns=tracker01.columns)
df
```

```python
tracker01["reception_date"]=pd.to_datetime(tracker01["receptiontimestamp"].dt.date)
tracker01.sort_values("reception_date")
```

```python
combo = df.merge(tracker01, how='left', left_on="date_counter", right_on="reception_date")
combo
```
